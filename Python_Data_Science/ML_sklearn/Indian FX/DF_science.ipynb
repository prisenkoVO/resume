SYMBOL

TOTALTRADES

Isin

```python
y = df.CLOSE
X = df.drop(['CLOSE'],axis = 1)
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 11)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 
print (N_train, N_test)
```

```python
%%time
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=5)
knn.fit(X_train, y_train)
print('Обучающая выборка: {:.2f}'.format(knn.score(X_train, y_train)))
print('Тестовая выборка: {:.2f}'.format(knn.score(X_test, y_test)))
```

```python
%%time
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
rfr.fit(X_train, y_train)
print('Обучающая выборка: {:.2f}'.format(rfr.score(X_train, y_train)))
print('Тестовая выборка: {:.2f}'.format(rfr.score(X_test, y_test)))
```
